```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
abalone_df = pd.read_csv("/content/abalone.csv")
```

```python
abalone_df.head()
```

```python
X_length_df = abalone_df['Length']
y = abalone_df['Rings']
```

```python
X_length_train, X_length_test, y_train, y_test = train_test_split(X_length_df, y, random_state=42)

X_length_train = pd.DataFrame(X_length_train)
y_train = pd.DataFrame(y_train)
X_length_test = pd.DataFrame(X_length_test)
y_test = pd.DataFrame(y_test)
```

```python
abalone_model = LinearRegression()
```

```python
abalone_model.fit(X_length_train, y_train) #arrays, dataframes,
```

```python
abalone_model.predict(X_length_test)
abalone_model.coef_
```

```python
abalone_model.score(X_length_test, y_test)
```

```python
import matplotlib.pyplot as plt

# Make predictions on the training data
y_train_pred = abalone_model.predict(X_length_train)

# Plot the training data and the regression line
plt.scatter(X_length_train, y_train, color='blue', label='Actual')  # Actual data points
plt.plot(X_length_train, y_train_pred, color='red', label='Predicted')  # Regression line
plt.xlabel('Length')
plt.ylabel('Rings')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```
