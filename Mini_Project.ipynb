**Data Preprocessing**

1. Handle Missing Values
2. Encoding the ```Sex``` and ```Embarked``` features
3. Selecting important features

````python
import pandas as pd
````

````python
titanic_df = pd.read_csv("/content/Titanic-Dataset.csv")
titanic_df
````

We can drop ```PassengerId```, ```Name```, ```Ticket``` and ```Cabin``` features

````python
titanic_df = titanic_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
````

````python
titanic_df
````

Next step is to handle missing values. In our case we will have to drop all rows with  NULL values

````python
titanic_df = titanic_df.dropna()
````

Let us check if there is any missing value in our columns/features

````python
titanic_df.isnull().sum()
````

Lets look at some characteristics of our data

````python
print(titanic_df.head())

print(titanic_df.describe())
````

Let us now encode the ```Sex``` , and ```Embarked``` features
We will use LabelEncoder

````python
from sklearn.preprocessing import LabelEncoder
````

````python
# Create a Label Encoder object
le = LabelEncoder()

# Fit and transform the 'Sex' column (Male -> 1, female -> 0)
titanic_df['Sex'] = le.fit_transform(titanic_df['Sex'])

# Fit and transform the 'Embarked' column (S -> 2, C -> 0, Q -> 1)
titanic_df['Embarked'] = le.fit_transform(titanic_df['Embarked'])
````
